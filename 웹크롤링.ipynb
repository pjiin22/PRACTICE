### beautifulsoup 활용

- 크롤링의 과정
    1. 파이썬으로 웹서버에 정보 요청 (개발자 도구)
    2. HTML데이터 가져오기
    3. 데이터 피싱
    4. 원하는 정보 저장

```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
keyword = '제주도'

url = f'https://search.naver.com/search.naver?ssc=tab.blog.all&sm=tab_jum&query={keyword}' #제주도를 서치할거라서

res = requests.get(url)
soup = bs(res.text, 'html.parser') #리퀘스트의 텍스트를 파싱해줘
```

```python
soup
```

```python
# 블로그들의 제목을 긁어오고 싶다면
## 개발자 모드에서 해당 부분의 class읽기

soup.find_all('a', class_='title_link')
```

```python
soup.find_all('a', class_='title_link')[0].text
```

```python
soup.find_all('a', class_='title_link')[0]['href'] #블로그 주소 가져오기
```

```python
## 리스트 컴프리핸션 사용

[i.text for i in soup.find_all('a', class_='title_link')]
```

```python
[i['href'] for i in soup.find_all('a', class_='title_link')]
```

```python
title = [i.text for i in soup.find_all('a', class_='title_link')]

len(title)
```

```python
# 날짜 긁어오기

date = [i.text for i in soup.find_all('span', class_='sub')]

len(date)
```

```python
# 내용 긁어오기

content = [i.text for i in soup.find_all('a', class_='dsc_link')]

len(content)
```

- 합치기

```python
df = pd.DataFrame({'title':title, 'date':date, 'content':content}) ##date길이가 왜 안맞는지 모르겠음
df
```
